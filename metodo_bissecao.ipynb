```python
f = lambda x: x**4 - x**3 - x**2 + x -3
```

```python
X=list()
Y=list()

for x in range(-3, 5, +1):
  X.append(x)
  Y.append(f(x))
  print(f'{x} index: {f(x)}')
```

```python
plt.plot(X, Y)
plt.show()
```

```python
def fParada(a, b, precisao):
  if(np.abs(a-b)<precisao):
    return True
  else:
    return False
```

```python
def funcao_bissecao(a,b,iter,precisao):
  fa = f(a)
  fb = f(b)
  if f(a)*f(b) >= 0:
    print("Método da Bisseção falhou!")
    return None  
  x = None
  i = 0
  criterio_de_precisao = False
  while((i<iter)|(criterio_de_precisao)):
    i = i + 1
    x = (a + b)/2
    fx = f(x)
    print(f'iter={i}, a={a}, b={b}, x={x}, f(a)={fa}, f(b)={fb}, f(x)={fx}')    
    if fa*fx < 0:        
       b = x
       fb = fx
    elif fb*fx < 0:
       a = x
       fa = fx
    elif fx == 0:
       print("Solução exata encontrada.")
       return x
    else:
       print("Método da Bisseção falhou!")
       return None
    criterio_de_precisao=fParada(a,b,precisao)
  return x
```

```python
a=-2
b=-1
interaction=100
precision=1e-06
x=funcao_bissecao(a,b,interaction,precision)
print(f'The result os bissection mode is:(x;f(x)) = ({x};{f(x)})')
```
